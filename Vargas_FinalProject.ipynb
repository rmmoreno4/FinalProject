### Questioning the Data 

#### Below is the visual data and my questions. In the beginning I imported all necessary code to make sure the csv file was able to be read. Each question has its own graph to read and inlcudes its own explanation.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
plt.rcParams["figure.figsize"] = (10, 6)
```

```python
superheroes = pd.read_csv("superhero_viz.csv", parse_dates = ["FIRST_YEAR"])
```

```python
superheroes.head()
```

### 1. How many DC characters have secret identities? How many Marvel characters have secret identities?

```python
identityType = superheroes["ID"].value_counts()
identityType.plot(kind = "pie" , fontsize= 8, autopct = "%1.0f%%")
plt.show()
```

```python
secret_iddc = pd.crosstab(superheroes["ID"] == "Secret Identity" , superheroes["UNIVERSE"] == "DC")
secret_iddc.index = ["Not Secret" , "Secret"]
secret_iddc.columns = ["Not DC" , "DC"]
secret_iddc
```

```python
secret_iddc = pd.crosstab(superheroes["ID"] == "Secret Identity" , superheroes["UNIVERSE"] == "DC")
secret_iddc.plot(kind = "bar")
plt.title("DC Secret Identities" , fontsize = 16 )
plt.show
```

```python
marvel_id = pd.crosstab(superheroes["ID"] == "Secret Identity" , superheroes ["UNIVERSE"] == "MARVEL")
marvel_id.index = ["Not Secret" , "Secret"]
marvel_id.columns = ["Not Marvel" , "Marvel"]
marvel_id
```

```python
marvel_id = pd.crosstab(superheroes["ID"] == "Secret Identity" , superheroes ["UNIVERSE"] == "MARVEL")
marvel_id.plot(kind = "bar")
plt.title("Marvel Secret Identities" , fontsize = 16)
plt.show
```

#### First I wanted to show the over all identity breakdown in the entire data set. Then, as we can see here there are more Marvel characters that have secret identities, DC has more false data which means it does not meet the criteria for the secret identities. I was curious to see which universe had more secret identities. The next few questions will explore diversity in female vs male characters in the universes as well.

### 2. What are the break downs of female and male characters in both universes as well?

```python
gender_universe = pd.crosstab(superheroes ["SEX"] , superheroes["UNIVERSE"])
gender_universe
```

```python
gender_universe.plot(kind = "bar")
plt.title("Types of Genders in Both Universes" , fontsize = 18)
plt.show()
```

#### I was surprised to see how big the gap was in the sex of Marvel characters. Marvel  clearly has more male than female characters. DC does as well, but the gap is not as wide. This was the most eye opening part of my research because though so many female characters have been introduced, there is still a huge gap between male and female characters overall. Here is where I start to think that there is still so much progress to be made when it comes to the representation of female characters in comics on a broader spectrum.

### 3. Are there more female or male characters with secret identities?

```python
sexID = pd.crosstab(superheroes["ID"], superheroes["SEX"])
sexID
```

```python
sexID = pd.crosstab(superheroes["ID"], superheroes["SEX"])
sexID.plot()
plt.title("Different ID Types by Gender")
plt.ylabel("Gender Type", fontsize = 10)
plt.xlabel("Identity Type", fontsize = 10)
plt.show()
```

#### Once I got through the first 2 questions it was no surprise that there were more males characters with secret identities because overall there are more male characters in general. The green line is our female characters and the pink is our male characters. It is very obvious that there is a huge difference in the two. There is also a vast difference in what type of indetities the male and female characters have. There is still so much more versatility for males than for females.
